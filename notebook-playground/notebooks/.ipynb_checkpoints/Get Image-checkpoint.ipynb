```python


landsat_item = gis.content.search('"Multispectral Landsat"', 'Imagery Layer')[0]
```

```python
landsat_item
```

```python
landsat = landsat_item.layers[0]
```

```python
def extract_stretch(bandids):
    return stretch(extract_band(landsat, bandids),
                   stretch_type='PercentClip',
                   min_percent=2, 
                   max_percent=2,
                   dra=True, 
                   gamma=[0.8,0.8,0.8])
```

```python
location["extent"]
```

```python


cambridge_gulf = extract_stretch([5, 4, 1])
cambridge_gulf.extent = {'xmin': 18.025980000000075,
 'ymin': -34.315089999999984,
 'xmax': 18.817980000000077,
 'ymax': -33.52308999999998, 'spatialReference': 4326}
cambridge_gulf

```
